```python
df = spark.read \
.format('csv') \
.option('header','true') \
.option('inferSchema','true') \
.load('/data/customers_100.csv')
```

```python
df.show(5)
```

```python
df.printSchema()
```

```python
df.filter('is_active = true').show()
```

```python
df.select('customer_id','name','city').filter('is_active=true').show()
```

```python
df.count()
```

```python
df.write\
.format('csv')\
.option('header','true')\
.save('/data/write_output.csv')
```
