```python
path = "gs://anshk-data/Data/1MB/customers.csv"
```

```python
rdd = spark.sparkContext.textFile(path)
```

```python
header = rdd.first()
```

```python
rdd_no_header = rdd.filter(lambda row:row!=header).map(lambda row:row.split(','))
```

```python
rdd_no_header.first()
```

```python
reduced_rdd = rdd_no_header.map(lambda row:(row[2],1)).reduceByKey(lambda x,y:x+y)
```

```python
reduced_rdd.collect()
```

```python
 group_rdd = rdd_no_header.map(lambda row:(row[2],1)).groupByKey()
```

```python
group_rdd.collect()
```

```python
group_rdd.map(lambda row:(row[0],len(row[1]))).collect()
```
