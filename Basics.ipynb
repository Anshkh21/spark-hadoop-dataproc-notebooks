```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName("WordCountDemo") \
.master("yarn") \
.getOrCreate()
```

```python
data = [
    "Apache Spark is fast",
    "Spark is easy to use",
    "Apache Spark supports batch processing",
    "Spark supports streaming and batch",
    "Big data processing with Spark",
    "Spark Spark Spark"
]
```

```python
rdd = spark.sparkContext.parallelize(data)
```

```python
print(rdd)
```

```python
rdd.collect()
```

```python
rdd_mapped = rdd.map(lambda line : line.split(" "))
```

```python
rdd_mapped.collect()
```

```python
words = rdd.flatMap(lambda line:line.split(" "))
```

```python
words.collect()
```

```python
word_map = words.map(lambda word:(word,1))
```

```python
word_map.collect()
```

```python
total = (word_map.reduceByKey(lambda a,b:a+b).sortByKey())
```

```python
total.collect()
```
